```python
import matplotlib.pyplot as plt
import seaborn as sns
import random as r
import time as t

L1=[]
for i in range(5000):
    L1.append(r.randint(1,100000))
a=t.time()
L1.sort()
b=t.time()
c=b-a

L2=[]
for i in range(10000):
    L2.append(r.randint(1,100000))
d=t.time()
L2.sort()
e=t.time()
f=e-d

L3=[]
for i in range(15000):
    L3.append(r.randint(1,100000))
g=t.time()
L3.sort()
h=t.time()
I=h-g

L4=[]
for i in range(20000):
    L4.append(r.randint(1,100000))
j=t.time()
L4.sort()
k=t.time()
l=k-j

L5=[]
for i in range(25000):
    L5.append(r.randint(1,100000))
m=t.time()
L5.sort()
n=t.time()
o=n-m

elements = [5000,10000,15000,20000,25000]
timetaken = [c,f,I,l,o]
print(timetaken)
figs=sns.lineplot(x=elements,y=timetaken)
figs.set_title('Time Taken vs No Of Elements in a List')
figs.set_ylabel('Time Taken')
figs.set_xlabel('No Of Elements in list')


```
