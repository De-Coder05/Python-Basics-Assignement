```python
def prime(n):
    for a in range(2,n//2 + 1):
        if n%a==0:
            flag=1
            break
        else:
            flag=0
    if flag==0:
        return 1
    else:
        return 0

import string as s
import random as r

print("Here are 100 random strings with lengths between 4 and 6: ")
for i in range(100):                                                       # (i)
    pwd=r.sample(s.ascii_letters,r.randint(4,6))
    print("".join(pwd))
print("-------x----------x---------x--------")

print("Here are all prime numbers between 600 and 800: ")
for i in range(600,800):                                                  # (ii)
    if prime(i)==1:
        print(i)
print("-------x----------x---------x--------")

print("Here are all numbers between 100 and 1000 that are divisible by 7 and 9: ")
for i in range(100,1000):                                                # (iii)
    if i%7==0 and i%9==0 :
        print(i)
```
